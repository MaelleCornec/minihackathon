## Visualisation des données Users

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des données

data = pd.read_csv("data/Users.txt", sep="\t")
data.head(15)
```

```python
print("The Users dataset has", data.shape[0], "rows and", data.shape[1], "columns.")
print("The columns are", data.columns.values)
```

```python
data.describe()
```

```python
data.describe(exclude=np.number)
```

```python
data.info()
```

```python
sns.displot(data, x="BirthYear", hue="Gender", multiple="stack")
```

```python
categorical_data = data.select_dtypes(include="object")

# Réalisation de graphe grâce sns.countplot pour toutes les variables catégorielles
for i in categorical_data.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(y=i, data=categorical_data)
    plt.show()
```
